The features in this dataset
The first column `respondent_id` is a unique and random identifier. The remaining 35 features are described below.

0 = Not at all concerned; 1 = Not very concerned; 2 = Somewhat concerned; 3 = Very concerned.

0 = No knowledge; 1 = A little knowledge; 2 = A lot of knowledge.

```python
# Read training dataset labels
df_train_labels = pd.read_csv('../data/raw/training_set_labels.csv')
df_train_labels.columns
```

```python
# Merge training data with labels
df_full = df_train.merge(df_train_labels, on='respondent_id', how='left')
df_full.columns
```

```python
# Choose six socio-demographic features
socio_demo_features = [
    'age_group',
    'sex',
    'education',
    'income_poverty',
    'race',
    'employment_status'
]

# Create a 23 subplot grid
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for ax, feature in zip(axes, socio_demo_features):
    # Compute normalized value counts (percentages)
    vc = (
        df_full[feature]
        .value_counts(dropna=False, normalize=True)
        .sort_index()
        * 100
    )
    
    # Plot bar chart
    vc.plot(kind='bar', ax=ax)
    
    # Styling
    ax.set_title(feature.replace('_', ' ').title(), fontsize=14)
    ax.set_ylabel("Percentage (%)")
    ax.set_xlabel("")
    ax.tick_params(axis='x', rotation=45)
    
    # Add % labels on bars
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height:.1f}%',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=9)

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

# Colors from your palette
FEMALE_COLOR = "#F08C66"   # Red
MALE_COLOR   = "#58ADC5"   # Blue
TEXT_COLOR   = "#000000"   # Black
BG_COLOR     = "#FFFFFF"   # White

# Rename and order age groups youngest  oldest
age_mapping = {
    '18 - 34 Years': '1834 y.o.',
    '35 - 44 Years': '3544 y.o.',
    '45 - 54 Years': '4554 y.o.',
    '55 - 64 Years': '5564 y.o.',
    '65+ Years': '65+ y.o.'
}

order = ['1834 y.o.', '3544 y.o.', '4554 y.o.', '5564 y.o.', '65+ y.o.']

df_plot = df_full.copy()
df_plot['age_group_clean'] = df_plot['age_group'].map(age_mapping)

# Compute counts by age  sex
grouped_counts = df_plot.groupby(['age_group_clean', 'sex']).size().unstack().reindex(order)

# Compute percentages within each age group
grouped_pct = grouped_counts.div(grouped_counts.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(12, 8), facecolor=BG_COLOR)
ax.set_facecolor(BG_COLOR)

# Plot stacked bars (raw counts)
bars_male = ax.bar(grouped_counts.index,
                   grouped_counts['Male'],
                   color=MALE_COLOR,
                   label='Male')

bars_female = ax.bar(grouped_counts.index,
                     grouped_counts['Female'],
                     bottom=grouped_counts['Male'],
                     color=FEMALE_COLOR,
                     label='Female')

# Add percentage labels inside bars
for i, age in enumerate(grouped_counts.index):
    male_h = grouped_counts.loc[age, 'Male']
    female_h = grouped_counts.loc[age, 'Female']
    total_h = male_h + female_h

    # Male percentage inside male segment
    if male_h > 0:
        ax.text(i, male_h / 2,
                f"{grouped_pct.loc[age, 'Male']:.1f}%",
                ha='center', va='center',
                color=TEXT_COLOR, fontsize=20)

    # Female percentage inside female segment
    if female_h > 0:
        ax.text(i, male_h + female_h / 2,
                f"{grouped_pct.loc[age, 'Female']:.1f}%",
                ha='center', va='center',
                color="white", fontsize=20)

# Titles & labels
ax.set_title("Distribution of Sex Across Age Groups", fontsize=30, color=TEXT_COLOR, pad=20)
ax.set_ylabel("Count", fontsize=20, color=TEXT_COLOR)
ax.set_xlabel("", fontsize=20)

# Tick formatting (remove zero, show K instead of 000)
def thousands_formatter(x, pos):
    if x >= 1000:
        return f"{int(x/1000)}K"
    elif x == 0:
        return ""   # remove 0 label
    return int(x)

ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

ax.tick_params(axis='y', labelsize=20, color=TEXT_COLOR)
ax.tick_params(axis='x', labelsize=20, color=TEXT_COLOR)

# Remove all axis spines except Y-axis for visibility
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(False)

# Legend below plot, no frame
legend = ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.12),
    ncol=2,
    fontsize=20,
    frameon=False
)

plt.tight_layout()
plt.show()

```

```python
COLOR = "#F08C66"     # Red
TEXT_COLOR = "#000000"
BG_COLOR = "#FFFFFF"

df_plot = df_full.copy()

# -------------------------------------------------
#  REPLACE LABELS
# -------------------------------------------------
df_plot['education'] = df_plot['education'].replace({
    "Missing": "N/A",
    "< 12 Years": "<12 Years",
    "College Graduate": "College Grad"
})

df_plot['income_poverty'] = df_plot['income_poverty'].replace({
    "Missing": "N/A",
    "<= $75,000, Above Poverty": "<=$75K",
    "> $75,000": ">$75K"
})

df_plot['employment_status'] = df_plot['employment_status'].replace({
    "Missing": "N/A",
    "Not in Labor Force": "Not in Labor"
})

df_plot['rent_or_own'] = df_plot['rent_or_own'].replace({
    "Missing": "N/A"
})

# -------------------------------------------------
#  YOUR REQUIRED CATEGORY ORDER (REAL  N/A)
# -------------------------------------------------
edu_order = ["College Grad", "Some College", "12 Years", "<12 Years", "N/A"]
income_order = [">$75K", "<=$75K", "Below Poverty", "N/A"]
employment_order = ["Employed", "Not in Labor", "Unemployed", "N/A"]
rent_order = ["Own", "Rent", "N/A"]

MAX_ROWS = 5
y_positions = [4, 3, 2, 1, 0]   # fixed top  bottom mapping


# -------------------------------------------------
#  PADDING FUNCTION
# -------------------------------------------------
def padded_series(col, order):
    order = order.copy()

    # add blanks to bottom until 5 total rows exist
    while len(order) < MAX_ROWS:
        order.append("")

    vc = col.value_counts(normalize=True) * 100
    vc = vc.reindex(order).fillna(0)

    return order, vc.values


# -------------------------------------------------
#  GENERATE LABELS + VALUES FOR EACH SUBPLOT
# -------------------------------------------------
edu_labels, edu_vals = padded_series(df_plot['education'], edu_order)
income_labels, income_vals = padded_series(df_plot['income_poverty'], income_order)
employment_labels, employment_vals = padded_series(df_plot['employment_status'], employment_order)
rent_labels, rent_vals = padded_series(df_plot['rent_or_own'], rent_order)

plots = [
    ("Education", edu_labels, edu_vals),
    ("Income", income_labels, income_vals),
    ("Employment", employment_labels, employment_vals),
    ("Home", rent_labels, rent_vals),
]


# -------------------------------------------------
#  FIGURE
# -------------------------------------------------
fig, axes = plt.subplots(
    2, 2,
    figsize=(14, 10.2),
    facecolor=BG_COLOR
)
axes = axes.flat

plt.subplots_adjust(hspace=0.28, wspace=0.12)

bar_height = 0.55


# -------------------------------------------------
#  PLOTTING LOOP
# -------------------------------------------------
for ax, (title, labels, vals) in zip(axes, plots):

    # bars
    ax.barh(y_positions, vals, color=COLOR, height=bar_height)

    # category labels drawn manually on left
    for y, label in zip(y_positions, labels):
        if label != "":
            ax.text(
                -1.5, y, label,
                ha='right', va='center',
                fontsize=14, color=TEXT_COLOR
            )

    # value labels inside bars
    for y, label, val in zip(y_positions, labels, vals):
        if label != "" and val > 0:
            ax.text(
                val / 2, y,
                f"{val:.1f}%",
                ha='center', va='center',
                fontsize=13, color="white", fontweight="bold"
            )

    ax.set_title(title, fontsize=20, color=TEXT_COLOR)

    # x-axis
    ax.tick_params(axis='x', labelsize=12)

    # remove y ticks completely
    ax.set_yticks([])

    # clean spines
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)


# -------------------------------------------------
#  SHARED X-AXIS LABEL
# -------------------------------------------------
# fig.text(0.5, 0.04, "%", ha='center', fontsize=18, color=TEXT_COLOR)

plt.tight_layout(rect=[0, 0.06, 1, 1])
plt.show()

```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd

COLOR_PALETTE = [
    "#F08C66",
    "#58ADC5",
    "#8AC6D6",
    "#CDE6EE",
]

LIGHT_COLORS = ["#58adc5", "#8ac6d6", "#cde6ee"]  # lowercase for matching

TEXT_COLOR = "#000000"
BG_COLOR = "#FFFFFF"

# Calculate race percentages
race_pct = df_train['race'].value_counts(normalize=True) * 100

# -----------------------------
# PIE CHART
# -----------------------------
fig, ax = plt.subplots(
    figsize=(6, 6),
    facecolor=BG_COLOR
)

wedges, texts, autotexts = ax.pie(
    race_pct.values,
    labels=race_pct.index,
    autopct="%.1f%%",
    startangle=90,
    counterclock=False,
    colors=COLOR_PALETTE[:len(race_pct)],
    textprops={"color": TEXT_COLOR, "fontsize": 16},
    wedgeprops={"edgecolor": "white"}
)

# Style percentage labels (conditional color)
for wedge, autotext in zip(wedges, autotexts):
    hex_color = mcolors.to_hex(wedge.get_facecolor())
    if hex_color.lower() in LIGHT_COLORS:
        autotext.set_color("black")
    else:
        autotext.set_color("white")

    autotext.set_fontweight("bold")
    autotext.set_fontsize(14)

# Title (slightly higher)
ax.set_title(
    "Race Distribution",
    fontsize=24,
    color=TEXT_COLOR,
    loc="center",
    pad=20
)

ax.axis("equal")  # ensure circle

plt.tight_layout()
plt.show()
```
